```python
%pylab inline
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
actividad=np.genfromtxt("./monthrg.dat")
```

```python
year=actividad[:,0]
mes=actividad[:,1]
media=actividad[:,3]
dev=actividad[:,4]
```

```python
yearxix=year[year>=1900]
mediaxix=media[year>=1900]
mesxix=mes[year>=1900]
devxix=dev[year>=1900]
```

```python
print yearxix
print mediaxix
print devxix
print mesxix
```

```python
tiempoxix=[]
for i in range(0,len(yearxix)):
    tiempoxix.append(yearxix[i]+((mesxix[i]-1)/12))
```

```python
manchas=plt.figure()
plt.plot(tiempoxix,mediaxix)
plt.xlabel("Time in years")
plt.ylabel("manchas solares/mes")
plt.title("Fluctuacion de manchas solares")
```
